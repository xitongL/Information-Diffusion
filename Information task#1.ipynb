# 1) if we use RA I's results as ground truth, what's the MAE (mean absolute error) of the machine's results?

```python
#find the difference by subtracting
Diff_1=Machine.subtract(RA_I)
Diff_1.head()
```

```python
#find absolute difference
Diff_1=Diff_1.abs()
Diff_1.head()
```

```python
#find MAE
Diff_1.mean()
Col_mean=pd.Series(Diff_1.mean())
Col_mean.mean()
```

```python
Diff_1.describe()
```

```python
Diff_1.mean().sort_values(ascending=False).plot(kind='bar',title='mean of abs difference machine vs. RA I')
```

```python
Diff_1.std().sort_values(ascending=False).plot(kind='bar',title='std of abs difference machine vs. RA I')
```

## How accurate the machine is?

```python
Machine_copy=Machine.copy()
Machine_copy
```

```python
def emotion_check(x):
    if x!=0:
        return 1
    else:
        return 0
```

```python
Machine_copy['Anger_check']=Machine_copy['Anger'].apply(emotion_check)
Machine_copy['Anxiety_check']=Machine_copy['Anxiety'].apply(emotion_check)
Machine_copy['Expect_check']=Machine_copy['Expect'].apply(emotion_check)
Machine_copy['Hate_check']=Machine_copy['Hate'].apply(emotion_check)
Machine_copy['Joy_check']=Machine_copy['Joy'].apply(emotion_check)
Machine_copy['Love_check']=Machine_copy['Love'].apply(emotion_check)
Machine_copy['Sorrow_check']=Machine_copy['Sorrow'].apply(emotion_check)
Machine_copy['Surprise_check']=Machine_copy['Surprise'].apply(emotion_check)
Machine_copy.drop(columns=['Anger','Anxiety','Expect','Hate','Joy','Love','Sorrow','Surprise'],inplace=True)
Machine_copy
```

```python
RA_I_copy=RA_I.copy()
```

```python
RA_I_copy['Anger_check']=RA_I_copy['Anger'].apply(emotion_check)
RA_I_copy['Anxiety_check']=RA_I_copy['Anxiety'].apply(emotion_check)
RA_I_copy['Expect_check']=RA_I_copy['Expect'].apply(emotion_check)
RA_I_copy['Hate_check']=RA_I_copy['Hate'].apply(emotion_check)
RA_I_copy['Joy_check']=RA_I_copy['Joy'].apply(emotion_check)
RA_I_copy['Love_check']=RA_I_copy['Love'].apply(emotion_check)
RA_I_copy['Sorrow_check']=RA_I_copy['Sorrow'].apply(emotion_check)
RA_I_copy['Surprise_check']=RA_I_copy['Surprise'].apply(emotion_check)
RA_I_copy.drop(columns=['Anger','Anxiety','Expect','Hate','Joy','Love','Sorrow','Surprise'],inplace=True)
RA_I_copy
```

```python
RA_J_copy=RA_J.copy()
```

```python
RA_J_copy['Anger_check']=RA_J_copy['Anger'].apply(emotion_check)
RA_J_copy['Anxiety_check']=RA_J_copy['Anxiety'].apply(emotion_check)
RA_J_copy['Expect_check']=RA_J_copy['Expect'].apply(emotion_check)
RA_J_copy['Hate_check']=RA_J_copy['Hate'].apply(emotion_check)
RA_J_copy['Joy_check']=RA_J_copy['Joy'].apply(emotion_check)
RA_J_copy['Love_check']=RA_J_copy['Love'].apply(emotion_check)
RA_J_copy['Sorrow_check']=RA_J_copy['Sorrow'].apply(emotion_check)
RA_J_copy['Surprise_check']=RA_J_copy['Surprise'].apply(emotion_check)
RA_J_copy.drop(columns=['Anger','Anxiety','Expect','Hate','Joy','Love','Sorrow','Surprise'],inplace=True)
RA_J_copy
```

## Metrics to measure the accurancy of the model

```python
from sklearn.metrics import f1_score
```

```python
# Machine vs. RA_I
f1_score(RA_I_copy,Machine_copy,average='macro')
```

```python
f1_score(RA_I_copy,Machine_copy, average='micro')
```

```python
f1_score(RA_I_copy, Machine_copy, average='weighted')
```

```python
# Machine vs. RA_J
f1_score(RA_J_copy,Machine_copy,average='macro')
```

```python
f1_score(RA_J_copy,Machine_copy,average='micro')
```

```python
f1_score(RA_J_copy,Machine_copy,average='weighted')
```

The F1 score can be interpreted as a weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. 

```python
from sklearn.metrics import log_loss
```

```python
log_loss(RA_I_copy,Machine_copy)
```

```python
log_loss(RA_J_copy,Machine_copy)
```

The goal of our machine learning models is to minimize this value. A perfect model would have a log loss of 0. Log loss increases as the predicted probability diverges from the actual label.

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(RA_I_copy,Machine_copy)
```

```python
accuracy_score(RA_J_copy,Machine_copy)
```

The best performance is 1, the larger the value is, the better the prediction
